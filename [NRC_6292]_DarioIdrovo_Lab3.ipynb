# Importación de librerias necesarias

```python
#Importación de las siguientes librerías:
import pandas as pd
import numpy as np

#Importar codificador de etiquetas
from sklearn import preprocessing
#Importar libreria de regresion lineal 
from sklearn.linear_model import LinearRegression
```

## Cargar el Dataset

```python
#Código para cargar el Dataset
detalles = pd.read_csv("CAR_DETAILS.csv")
detalles.head(10)
```

```python
detalles.describe()
```

```python
# objeto label_encoder sabe cómo entender las etiquetas de palabras
label_encoder = preprocessing.LabelEncoder()

# Codificar etiquetas en la columna 'Car_Name', 
# esta nueva etiqueta se denomina 'Car_Types_Name' y se la guarada en una nueva columna
detalles['Car_Name_Trans'] = label_encoder.fit_transform(detalles['Car_Name'])
detalles['Fuel_Type_Trans'] = label_encoder.fit_transform(detalles['Fuel_Type'])
detalles['Seller_Type_Trans'] = label_encoder.fit_transform(detalles['Seller_Type'])
detalles['Transmission_Trans'] = label_encoder.fit_transform(detalles['Transmission'])
detalles.head(10)
```

## Preprocesamiento del dataset. Transformaciones previas necesarias para la modelación

```python
#Variables objetivos y variables de entrada
VariableObjetivo=['Selling_Price']
VariablesEntrada=['Car_Name_Trans', 'Year', 'Present_Price', 'Kms_Driven', 'Fuel_Type_Trans', 'Seller_Type_Trans', 'Transmission_Trans', 'Owner']

#Transformación de las caracteristicas
X=detalles[VariablesEntrada].values
Y=detalles[VariableObjetivo].values
```

```python
#Estandarización de los datos
from sklearn.preprocessing import StandardScaler
VariablesEntrada_Scaler=StandardScaler()
VariableObjetivo_Scaler=StandardScaler()
```

```python
#Almacenamiento del objeto de ajuste para referencia posterior
VariablesEntrada_Fit=VariablesEntrada_Scaler.fit(X)
VariableObjetivo_Fit=VariableObjetivo_Scaler.fit(Y)
```

```python
#Generación de los valores estandarizados de X e y
X=VariablesEntrada_Fit.transform(X)
Y=VariableObjetivo_Fit.transform(Y)
```

## División del dataset en datos de entrenamiento y datos de test 

```python
#Division de los datos en un conjunto de entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
#Comprobación rápida del tamaño de los conjuntos de datos de entrenamiento y prueba
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

## Regresión Lineal

```python
#Codo de modelo
detalles_r_lineal = LinearRegression()
#Se imprime los parametros de la regrsion lineal
print(detalles_r_lineal)
```

```python
#Se crea el modelo de los datos de entrenamiento
#Uso del modelo, llamando al .fit()
reg_lineal = detalles_r_lineal.fit(X_train,Y_train)
prediction=reg_lineal.predict(X_test)

#Se obtienen resultados con .score()
resul = detalles.score(X_train,Y_train)
#Con el .score usado anteriormente se obtienen los coeficientes de determinación
print(f"Coeficiente de determinación: {resul}")

#Representacion del coeficiente con .intercept_ b0
print(f"Coeficiente Intercepta: {detalles.intercept_}")

#Representacion del coeficiente que representa b1
print(f"Coeficiente Pendiente: {detalles.coef_}")
```

```python
#Se crea el modelo de los datos de test
#Uso del modelo, llamando al .fit()
reg_lineal = detalles_r_lineal.fit(X_test,Y_test)
prediction=reg_lineal.predict(X_train)

#Se obtienen resultados con .score()
resul = detalles.score(X_test,Y_test)
#Con el .score usado anteriormente se obtienen los coeficientes de determinación
print(f"Coeficiente de determinación: {resul}")

#Representacion del coeficiente con .intercept_ b0
print(f"Coeficiente Intercepta: {detalles.intercept_}")

#Representacion del coeficiente que representa b1
print(f"Coeficiente Pendiente: {detalles.coef_}")
```
